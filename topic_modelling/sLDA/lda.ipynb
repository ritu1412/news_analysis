```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = [2, 5, 3]
n_samples = 1000

# Generate samples
samples = np.random.dirichlet(alpha, n_samples)

# Plot samples
plt.figure(figsize=(8, 8))
plt.scatter(samples[:, 0], samples[:, 1], marker='o', s=10, alpha=0.6, color='blue')
plt.xlabel('Topic 1 Proportion')
plt.ylabel('Topic 2 Proportion')
plt.title('Samples from a Dirichlet Distribution with =(2, 2, 2)')
plt.show()
```
